```python
### Bar Chart â€“ Subscription Type Distribution -

import matplotlib.pyplot as plt

df['Subscription Type'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Subscription Type Distribution')

plt.xlabel('Subscription Type')
plt.ylabel('Number of Users')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## Pie Chart - Gender Distribution

explode= [0, 0, 0.1]
          
df['Gender'].value_counts().plot(kind='pie', 
                                 autopct='%1.1f%%', 
                                 startangle=90, 
                                 shadow=True, 
                                 explode=explode, 
                                 colors=['#e83f16', '#e116e8', '#16e8b8'],
                                 wedgeprops={'edgecolor': '#08110f', 'linewidth': 1.7, 'linestyle': '-', 'fill':True}
                                 )

plt.title('Gender Distribution')
plt.ylabel('')  # Hiding y-label
plt.tight_layout()
plt.show()
```

```python
## Histtogram - Daily Watch Time Distribution -

import numpy as np

bins = np.arange(0, 5, 0.5) 

plt.hist(df['Daily Watch Time'], bins=10, edgecolor='black', color='orange')

plt.title('Daily Watch Time Distribution')

plt.xlabel('Hours Watched Per Day')
plt.ylabel('Number of Users')

plt.xticks(bins)

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
## Box Ploting â€“ Age by Subscription

import seaborn as sns

sns.boxplot(data=df, 
            x='Subscription Type', 
            y='Age', 
            palette='pastel'
           )

plt.title('Age Distrribution by Subscription')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
## bar Chart - 

df.groupby(['Country', 'Is Active']).size().unstack().plot(kind='bar', 
                                                           stacked=True, 
                                                           colormap='coolwarm'
                                                          )
# Note:- 0-Inactive & 1-Active.

plt.title('Active vs Inactive Users by Country')

plt.ylabel('User Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
