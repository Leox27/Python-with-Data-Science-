```python
import pandas as pd
```

```python
df = pd.read_csv("Book4.csv")
df                          # Note: if you use "print(df)" it gives you in text format
```

```python
df.isnull().copy()
```

```python
df.isnull().sum() # column wise missing data count
```

```python
df.dropna(axis=0) # Drops row wise missing value
                  # We can use df.dropna()
```

```python
df.dropna(axis=1) # # Drops column wise missing value
```

```python
df.fillna(0) # Fill "0" in emty cell
```

```python
# Fill mean in empty cells by using mean()

df["Age"].fillna(df["Age"].mean())
```

```python
df["Age"].fillna(df["Age"].mean()).round(2) # Used .round() for 2-decimal
```

```python
# Forward fill 
df.fillna(method="ffill")
```

```python
# Backword fill 
df.fillna(method="bfill")
```

```python
## Latest method

# Forward fill 
df.ffill()
```

```python
# Backward fill 
df.bfill()
```

```python
### String methods in Pandas

# .str.contain()
df["City"].str.contains("Delhi", case=False) # Using to see that the "Delhi" contain in"City" or not
```

```python
### Apply funs - Creates new gruop or column in reference of one

# .apply()
df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x >= 25 else "Minor")
df2
```

```python
def isminor(x): # By using the fun 
    return "Adult" if x>=25 else "Minor"

df2["Age Group"] = df2["Age"].apply(isminor)
df2
```
