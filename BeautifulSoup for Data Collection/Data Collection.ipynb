```python
from bs4 import BeautifulSoup
```

```python
### Making a BeautifulSoup Object: -

# Choose a file
with open("htmls/Page1.html") as f:
    content = f.read()

# Making a soup
soup = BeautifulSoup(content, "html.parser")
h3s = soup.find_all("h3")
h3s
```

```python
# Now we find the titles 

for h3 in h3s:
    print(h3.find("a")['title'])
```

```python
# It gives all code in article tag
articles = soup.select("article.product_pod")
articles
```

```python
# Collect all titles and the proce of the books

list = []

for article in articles:
    title = article.find("h3").find("a")["title"]
    # print(title) # Total 20 title of books in 1st page 😂 
    
    price = article.select_one("p.price_color").text.split("£")[1] # Splitting the text "£".
    # print(price) # price of the each book

    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    # print(classes)
    
    list.append([title, price, rating])
    
list # In text form
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list, columns=['Book Title', 'Price', 'Classes'])
```
