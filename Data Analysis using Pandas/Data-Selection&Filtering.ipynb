```python
### .at[] & .iat[] methods

df.at[0, "Actor"] # Used for only locating single cell in dataset with string
```

```python
df.iat[0, 0] # method follow only indices
```

```python
### Filtering with logical conditions

df[df["IMDb"] > 6] # For single column
```

```python
df[(df["IMDb"] > 6) & (df["Year"] > 2018)] # Filtering woth two columns using "&" operator 
```

```python
df[(df["IMDb"] > 6) | (df["Year"] > 2018)] # Filtering woth two columns using "|" operator 
```

```python
# By using ".query()" method

df.query("Year > 2018 and IMDb > 7") 
```

```python
# Dynamic columns method using f-string

col = "IMDb"
df.query(f"{col} > 6 ")
```

```python
# Chained comparison

df.query("6 < IMDb < 8.2") 
```
