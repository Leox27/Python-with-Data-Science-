```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris_data.xlsx")
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
x = data.iloc[ : , :-1]
```

```python
y = data.iloc[ : , -1]
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x, y)
```

```python
predictions = model.predict([[5.9, 3.2,	4.8, 1.8]])
predictions
```

```python
data_test = pd.read_excel("iris-test.xlsx")
```

```python
x_test = data_test.iloc[ :, :-1]
x_test
```

```python
y_test = data_test.iloc[ :, -1]
y_test
```

```python
prediction_test = model.predict(x_test)
```

```python
prediction_test
```

```python
actual = y_test.values # Confirming that is prediction are right !
actual
# Right 😂 
```

```python
# Checking Accuracy of the model -

count = 0
for i in range(0, len(actual)):
    if actual[i] == prediction_test[i]: # matching actual data and preducted data
        count += 1
    
print(((count*100)/len(actual)))
```

```python
# Wow it's 100 % Accuracy 😂 
```
