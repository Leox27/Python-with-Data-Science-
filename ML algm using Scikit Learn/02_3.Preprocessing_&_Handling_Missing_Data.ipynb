```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df["income_cat"] = pd.cut(df["median_income"], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels = [1, 2, 3, 4, 5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.3, random_state = 42) # 'random_state' is seed

for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
for sett in (strat_train_set, strat_test_set):
    sett.drop("income_cat", axis=1, inplace=True)
```

```python
df  = strat_train_set.copy()
```

```python
housing = df.drop(["median_house_value"], axis=1)
housing_labels = df["median_house_value"].copy()
```

```python
### Fillling  null data by Scikit Learn -

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = "most_frequent") # Only for numbers not strings
                                ### Other available strategies:
                                
                                # 1."mean" – replaces with mean value
                                # 2."most_frequent" – for the most common value (can handle categorical)
                                # 3."constant" – fill with a fixed value using * fill_value =...

housing_num = housing.select_dtypes(include = [np.number, 'object'])
```

```python
imputer.fit(housing_num) # fit the data in model
```

```python
# getting mean 

imputer.statistics_
```
