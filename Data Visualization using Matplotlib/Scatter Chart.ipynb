```python
import matplotlib.pyplot as plt
```

```python
# Basic Scatter Plot -

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]

plt.scatter(study_hours, exam_scores)

plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
### colors = ['red' if score < 45 else 'green' for score in exam_scores]


study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 95, 55, 80, 45, 75, 85, 90]

sizes = [score * 2 for score in exam_scores]
colors = ['red' if score < 45 else 'green' for score in exam_scores]

plt.scatter(study_hours, exam_scores, s=sizes, c=colors)

plt.title('Colored with Size Scatter Plot 👩 ')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
### Colored with Size Scatter Plot -

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 95, 55, 80, 45, 75, 85, 90]

sizes = [score * 2 for score in exam_scores]
scores = [23, 45, 67, 56, 34, 78, 89, 90, 34]

plt.scatter(study_hours, exam_scores, s=sizes, c=scores, cmap='viridis') #cmap for color-bar

plt.colorbar(label='Score') # Color-Bar 
# Link:- https://matplotlib.org/stable/users/explain/colors/colormaps.html

plt.title('Colored with Size Scatter Plot 👩')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
### Adding Annotations -

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 95, 55, 80, 45, 75, 85, 90]

sizes = [score * 2 for score in exam_scores]
scores = [123, 215, 367, 456, 534, 678, 789, 590, 934]

plt.scatter(study_hours, exam_scores, s=sizes, c=scores, cmap='viridis')

plt.colorbar(label='Score') # Color-Bar 

plt.scatter(study_hours, exam_scores) # Plotting 

# Adding Annotions on score
for i in range(len(study_hours)):
    plt.annotate(f'S-{i+1}', (study_hours[i] - 0.25, exam_scores[i] - 5)) # Adjusting th positions

plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
### Multiple Groups in One Plot -

import numpy as np

# Assume - Class A 
class_a_hours = [2, 4, 6, 8, 10]
class_a_scores = [45, 55, 65, 85, 90]

# Assume - Class B
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
