```python
### Comparing Virat Kohli and Virender Sehwag plot chart
# Multiple Lines in One Plot

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100] # Runs of Virat Kohli
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0] # Runs of Virender Sewwag

plt.plot(years, kohli, label="Virat Kohli") # Using "label= " for denoting plot string of runs wrt-year
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison of V.K.& V.S.")
plt.legend()
plt.show()
```

```python
# Don't use the legend() instead use the lebel()
# Bad Practice

plt.plot(years, kohli)
plt.plot(years, sehwag)

plt.legend(["Kohli", "Sehwag"])  # prone to mismatch

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison of V.K.& V.S.")

plt.show()
```

```python
### Comparing Virat Kohli and Virender Sehwag plot chart
# Using Format Strings

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

kohli = [0, 100, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100] # Runs of Virat Kohli
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0] # Runs of Virender Sewwag
Dhoni = [0, 100, 200, 300, 400, 500, 600, 700, 800, 1000] # Runs of Dhoni
Sharma = [0, 200, 400, 600, 800, 1000, 1200, 1300, 400, 200] # Runs of Sharma

plt.plot(years, kohli, 'r|--', label="Virat Kohli") # Using "label= " for denoting plot string of runs wrt-year
plt.plot(years, sehwag, 'g_:', label="Virender Sehwag")
plt.plot(years, Dhoni, 'yd:', label="Dhoni")
plt.plot(years, Sharma, 'b4-', label="Sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison of V.K.& V.S.")
plt.legend()
plt.show()
```

```python
### Colour and line string arguments

plt.plot(years, kohli, color='orange', linestyle='--', linewidth=4, label="Kohli") # With line-width
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison of V.K.& V.S.")

plt.legend()
plt.show()
```

```python
### methods()

plt.plot(years, kohli, color='orange', linestyle='--', linewidth=4, label="Kohli") # With line-width
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison of V.K.& V.S.")

plt.grid(True)
plt.tight_layout() # IT will show on in .py file
plt.legend()
plt.show()
```
