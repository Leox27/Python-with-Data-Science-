```python
import matplotlib.pyplot as plt
```

```python
### Bar Chart - Basic

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")

# plt.show() # No need to use this fun bcz this used in .py file to visua;ize the chart
```

```python
### Setting Bar Width & Side-by-Side Bar Charts

# Vertical Bar Chart -
import numpy as np # Used for baar positions

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(x - width, sachin, color='orange', width=width, label="Sachin") # 1-0.25 = 0.75, which means Sachin's bar is 0.25 from left.
plt.bar(x, sehwag, width=width, color='gray', label="Sehwag") # middle of the x-position 0.125 from left & 0.125 from right.
plt.bar(x + width, kohli, width=width, color='green', label="Kohli") # 1+0.25 = 0.75, which means Kohli's bar is 0.25 from right.

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")

plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()

# plt.show()
```

```python
### If the bars were 4 😂

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
dravid = [100, 300, 800, 1100, 1300, 1000, 1400, 1000, 700, 400]

x = np.arange(len(years)) # numpy arange()
width = 0.2

plt.bar(x - 1.5 * width, sachin, width=width, label="Sachin")
plt.bar(x - 0.5 * width, sehwag, width=width, label="Sehwag")
plt.bar(x + 0.5 * width, kohli, width=width, label="Kohli")
plt.bar(x + 1.5 * width, dravid, width=width, label="Dravid")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Comparison of 4 players")

plt.xticks(x, years)
plt.legend()
plt.tight_layout()
# plt.show()
```

```python
# Horizontal Bar Chart - plt.barh()

players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.ylabel("Players")
plt.title("First 5-Year Performance of Indian Batsmen")

plt.tight_layout()
# plt.show()
```

```python
### If the horizaontal bars were 4 😂 Note: Self
# I do bcz most of the these type of info visualzed by horizontal bar.

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
dravid = [100, 300, 800, 1100, 1300, 1000, 1400, 1000, 700, 400]

y = np.arange(len(years)) # numpy arange()
height = 0.2

plt.barh(y - 1.5 * height, sachin, height=height, label="Sachin")
plt.barh(y - 0.5 * height, sehwag, height=height, label="Sehwag")
plt.barh(y + 0.5 * height, kohli, height=height, label="Kohli")
plt.barh(y + 1.5 * height, dravid, height=height, label="Dravid")

plt.xlabel("Runs Scored")
plt.ylabel("Years")
plt.title("Comparison of 4 players")

plt.yticks(y, years) # Replaces index with labels
plt.legend()
plt.tight_layout()
# plt.show()
```

```python
### plt.text() : Text on bars - 

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="#FFD700") # We can even use the color-code(hex color)

# Adding label on top of bar
for r in range(len(players)): # 'r' is runs.
    plt.text(r, runs[r] + 10, str(runs[r]), ha='center') # r is the index
                                                         # +10 is the distance of runs above the bar
                                                         # str(runs[r]) is print the runs
                                                         # ha='center' is alignment.

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
# plt.show()
```
