```python
### axvline - 

import numpy as np # Using NumPy ☻ 

ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

# bins = [10, 20, 30, 35, 40, 45, 50, 60, 70]
bins = np.arange(0, 71, 10) # bins by using 'arange()'

plt.hist(ages, bins=bins, edgecolor='black') 

plt.axvline(27, color='gold', linestyle='--', linewidth=0.7, label='avg') # See in Histogram the line at age 27

plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

```

```python
import numpy as np # Using NumPy ☻ 

ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

# bins = [10, 20, 30, 35, 40, 45, 50, 60, 70]
bins = np.arange(0, max(ages)+10, 10) # used by 'arange()'

plt.hist(ages, bins='auto', edgecolor='black') # Auto bins

plt.axvline(25, color='gold', linestyle='--', linewidth=2.7, label='avg', alpha=0.5) # See in Histogram the line at age 27

plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

```python
### mean() using numpy

ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]

# bins = [10, 20, 30, 35, 40, 45, 50, 60, 70]
bins = np.arange(0, max(ages)+10, 10) # used by 'arange()'
mean_age = np.mean(ages) 

plt.hist(ages, bins='auto', edgecolor='black') # Auto bins

plt.axvline(25, color='gold', linestyle='--', linewidth=2.7, label='avg', alpha=0.5) # See in Histogram the line at age 27
plt.axvline(mean_age, color='red', linestyle='--', label=f'Mean: {mean_age:.1f}') # We can add multiple axvlines
                                                 # label=f'Mean: {mean_age:.1f}' - in label the value of mean

plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```
