```python
import matplotlib.pyplot as plt
```

```python
### basic Subplot: Implementing multiple plots

x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x] # 1st plot
y2 = [i ** 2 for i in x] # 2nd plot

# 1 row and 2 columns fpr 1st plot
plt.subplot(1, 2, 1)  # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title('Double of x')

# 1 row and 2 columns fpr 2st plot
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Square of x')

plt.tight_layout() # For accurate reading
plt.show()
```

```python
### 22 Grid Subplots -

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [i * 2 for i in x] # 1st plot
y2 = [i ** 2 for i in x] # 2nd plot
y3 = [i ** 0.5 for i in x] # 3rd plot
y4 = [10 - i for i in x] # 4th plot

plt.figure(figsize=(8, 6))  # Optional: make it bigger

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')

plt.tight_layout()
plt.show()
```

```python
### Accessing in subplots - For individual plots

fig, axs = plt.subplots(1, 2, figsize=(10, 4)) # 1-row & 2-columns - Font size

axs[0].plot(x, y1) # 1st plot 
axs[0].set_title('x * 2')

axs[1].plot(x, y2) # 2nd plot
axs[1].set_title('x squared')

# Using the fig.
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.90)  # So, Title does'nt overlap

fig.savefig('my_plots.png') # Save image

plt.show()
```

```python
### Looping over subpots

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]

fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

# By looping for 2x2 grid plot
for i in range(2):
    for j in range(2):
        
        idx = i * 2 + j
        
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
```
