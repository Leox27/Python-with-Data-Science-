```python
import matplotlib.pyplot as plt
```

```python
### Basic Pie Chart -

labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

colors = ['#e9240c','#e9df0c','#13e90c','#d80ce9'] # Custom HEX-colors from HTML-color code

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
### Edgedrop: Add edges and style slices

plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': '#0cd5e9', 'linewidth': 2.7, 'linestyle': '--', 'fill':True} # If 'fill':False then all color will blank
)
plt.show()
```

```python
### Explode
explode = [0.1, 0, 0, 0.2] # Sachin's & Yuvraj's Pizza piece will be detached 😂 

plt.title("Explode Pie Chart")
plt.pie(runs,
    labels=labels,
    colors=colors,
    explode=explode,
    wedgeprops={'edgecolor': '#0cd5e9', 'linewidth': 2.7, 'fill':True} # If 'fill':False then all color will blank
)
plt.show()
```

```python
### Shadow effect

explode = [0.1, 0, 0, 0.2] # Sachin's & Yuvraj's Pizza piece will be detached 😂 

plt.title("Explode Pie Chart")
plt.pie(runs,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True, # Shadow is True
    wedgeprops={'edgecolor': '#0cd5e9', 'linewidth': 2.7, 'fill':True}, # If 'fill':False then all color will blank
    startangle=270 # Rotate the Pie Chart
)
plt.show()
```

```python
###  Percentage show

explode = [0.1, 0, 0, 0.2] # Sachin's & Yuvraj's Pizza piece will be detached 😂 

plt.title("Explode Pie Chart")
plt.pie(runs,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True, # Shadow is True
    wedgeprops={'edgecolor': '#0cd5e9', 'linewidth': 2.7, 'fill':True}, # If 'fill':False then all color will blank
    startangle=270, # Rotate the Pie Chart
    autopct='%1.1f%%'
)
plt.show()
```

```python
### Crowded Pie Chart
# Avoid this Pie Chart is the categories are far OR use "others" word and add to it other categories

languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP', 'SQL']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1, 0.5]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```

```python
# Soln on Crowded Charts: Use 'others'

languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Others']
usage = [30, 20, 10, 10, 7, 15.5]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Soln)")
plt.show()
```
